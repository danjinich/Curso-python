### Graficas de distribuci√≥n
https://seaborn.pydata.org/generated/seaborn.distplot.html

```python
sns.distplot(titanic['age'].dropna())
```

```python
sns.distplot(titanic['age'].dropna(), bins=40, 
             hist=True, color="g",kde=True, 
             rug=True, vertical=True)
```

```python
sns.distplot(titanic['age'].dropna(), rug=True, 
             bins=15, vertical=True, rug_kws={"color": "k"},
             kde_kws={"color": "k", "lw": 3, "label": "KDE", 
                      "alpha":1, "lw":2},
             hist_kws={"histtype": "barstacked", "linewidth": 5, 
                       "alpha": .5, "color": "crimson", 
                       "label":"Hist"})
```

```python
bins = np.linspace(0, 90, 13)
sns.set(style="darkgrid")
sns.distplot(titanic['age'].dropna(), rug=False, 
             bins=bins, vertical=True, 
             kde_kws={"color": "k", "lw": 3, "label": "KDE", 
                      "alpha":.5, "lw":2},
             hist_kws={"histtype": "bar", "linewidth": 5, 
                       "alpha": .5, "color": "crimson", 
                       "label":"Hist"})
```

```python
f = plt.figure(figsize=(6, 6))
gs = f.add_gridspec(2, 2)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[0, 0])
    sns.distplot(titanic['age'].dropna(), bins=10, 
                 hist=True, color="red",kde=True, 
                 rug=True, vertical=False, ax=ax)

with sns.axes_style("white"):
    ax = f.add_subplot(gs[0, 1])
    sns.distplot(titanic['age'].dropna(), bins=10, 
                 hist=True, color="red",kde=True, 
                 rug=True, vertical=False, ax=ax)

with sns.axes_style("ticks"):
    ax = f.add_subplot(gs[1, 0])
    sns.distplot(titanic['age'].dropna(), bins=10, 
                 hist=True, color="red",kde=True, 
                 rug=True, vertical=False, ax=ax)

with sns.axes_style("whitegrid"):
    ax = f.add_subplot(gs[1, 1])
    sns.distplot(titanic['age'].dropna(), bins=10, 
                 hist=True, color="red",kde=True, 
                 rug=True, vertical=False, ax=ax)

f.tight_layout()
```

#### Paletas de color
https://chrisalbon.com/python/data_visualization/seaborn_color_palettes/

```python
#Secuencial
sns.palplot(sns.color_palette("Paired", 10))
```

```python
#Continua
sns.palplot(sns.color_palette("YlGnBu", 10))
```

```python
#Divergente
sns.palplot(sns.color_palette("RdYlGn", 10))
```

```python
f = plt.figure(figsize=(6, 6))
gs = f.add_gridspec(2, 2)

ax = f.add_subplot(gs[0, 0])
sns.countplot(x="class", hue="who", 
              data=titanic)

with sns.color_palette("YlGnBu"):
    ax = f.add_subplot(gs[0, 1])
    sns.countplot(x="class", hue="who", 
                  data=titanic)

with sns.color_palette("RdYlGn"):
    ax = f.add_subplot(gs[1, 0])
    sns.countplot(x="class", hue="who", 
                  data=titanic)

with sns.color_palette("Paired"):
    ax = f.add_subplot(gs[1, 1])
    sns.countplot(x="class", hue="who", 
                  data=titanic)
    

f.tight_layout()
```

### Regresiones
https://seaborn.pydata.org/generated/seaborn.lmplot.html

```python
g=sns.lmplot(x="total_bill",                    
           y="percent",                    
           data=tip)
g = (g.set_axis_labels("Cuenta total", "Porcentage").set(xlim=(0,60),ylim=(0,1)))
plt.title("Propina pagada por costo de cuenta")
plt.show(g)
```

```python
g=sns.lmplot(x="total_bill",                    
           y="percent",                    
           data=tip,
            aspect=4, 
            order=2)
g = (g.set_axis_labels("Porcentage","Cuenta total").set(xlim=(0,60),ylim=(0,1)))
plt.title("Propina pagada por costo de cuenta")
plt.show(g)
```

```python
sns.jointplot(x="total_bill",                    
              y="percent",                    
              data=tip, 
              kind="reg")
```

```python
sns.lmplot(x="sepal_width", y="petal_width", 
           data=iris, aspect=3)
```

### Analisis rapido

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```

```python
g = sns.PairGrid(iris, hue="species")
g = g.map_upper(plt.scatter)
g = g.map_diag(plt.hist, histtype="step", linewidth=3)
g = g.map_lower(plt.scatter)
g = g.add_legend()
```

```python
g = sns.PairGrid(iris)
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.kdeplot, lw=2)
```
