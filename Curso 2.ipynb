```python
df=pd.DataFrame({'palabras':ili,'num':len_p})
```

```python
df2=pd.DataFrame({'palabras':s_ili,'num':s_len_p})
```

## Matplotlib con Seaborn
Es una biblioteca para graficar, se utilisa mucho en relación con pandas. 
Graficas posibles: https://seaborn.pydata.org/examples/index.html
Colores: https://seaborn.pydata.org/tutorial/color_palettes.html

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Vamos a utilizar la misma BD que en la seccion anterior.

```python
fifa=pd.read_csv('fifa_ranking.csv')
```

```python
fifa['rank_date']=fifa['rank_date'].astype('datetime64')
```

```python
fifa['year']=fifa['rank_date'].apply(lambda x: x.year)
```

```python
fifa.head(10)
```

### Graficas basicas

### Regresiones

```python
g=sns.lmplot(x="year",                    
           y="rank",                    
           data=fifa[fifa['country_abrv']=='MEX'])
g = (g.set_axis_labels("Año","Rango").set(xlim=(1990,2020),ylim=(0,50)))
plt.title("Rango de México")
plt.show(g)
```

```python
g=sns.lmplot(x="year",                    
           y="rank",                    
           data=fifa[fifa['country_abrv']=='MEX'],                    
           aspect=4, 
            order=3)
g = (g.set_axis_labels("Año","Rango").set(xlim=(1990,2020),ylim=(0,50)))
plt.title("Rango de México")
plt.show(g)
```

```python
sns.jointplot(x="year", 
              y="rank", 
              data=fifa[fifa['country_abrv']=='MEX'], 
              kind="reg")
```

```python
paises=fifa[(fifa['country_abrv']=='GER') |
           (fifa['country_abrv']=='ITA') |
           (fifa['country_abrv']=='ARG') |
           (fifa['country_abrv']=='BRA')]
```

Se pueden cambiar los marcadores, a cuales en https://matplotlib.org/3.1.3/api/markers_api.html

```python
sns.lmplot(x="year", y="rank", hue="country_abrv", data=paises, aspect=4, markers=['o','x','s','v'])
```

```python
sns.lmplot(x="year", y="rank", hue="country_abrv",
           row="confederation", data=paises, aspect=4);
```

### Categoricas

```python
f, ax = plt.subplots()
sns.boxplot(x="country_full", y="rank", data=paises)
sns.despine(offset=10, trim=True);
```

```python
f, ax = plt.subplots()
sns.violinplot(x="country_full", y="rank", data=paises)
sns.despine(offset=10, trim=True);
```

```python
sns.pairplot(paises[['total_points','rank_change','year','previous_points']])
```

```python
sns.pairplot(paises[['total_points','rank_change','year','previous_points', 'country_abrv']], hue='country_abrv')
```

```python
g=sns.PairGrid(paises[['total_points','rank_change','year','previous_points']])
g.map_upper(sns.scatterplot)
g.map_lower(sns.barplot)
g.map_diag(sns.kdeplot)
```

### Diseño

```python
f = plt.figure(figsize=(6, 6))
gs = f.add_gridspec(2, 2)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[0, 0])
    sns.violinplot(x="country_full", y="rank", data=paises, ax=ax)

with sns.axes_style("white"):
    ax = f.add_subplot(gs[0, 1])
    sns.violinplot(x="country_full", y="rank", data=paises, ax=ax)

with sns.axes_style("ticks"):
    ax = f.add_subplot(gs[1, 0])
    sns.violinplot(x="country_full", y="rank", data=paises, ax=ax)

with sns.axes_style("whitegrid"):
    ax = f.add_subplot(gs[1, 1])
    sns.violinplot(x="country_full", y="rank", data=paises, ax=ax)

f.tight_layout()
```

```python
sns.palplot(sns.color_palette("Paired", 10))
```

```python
sns.palplot(sns.diverging_palette(220, 20, n=10))
```

```python
sns.palplot(sns.cubehelix_palette(10))
```

```python
f = plt.figure(figsize=(6, 6))
gs = f.add_gridspec(2, 2)

ax = f.add_subplot(gs[0, 0])
sns.violinplot(x="country_full", y="rank", data=paises, ax=ax)

with sns.cubehelix_palette():
    ax = f.add_subplot(gs[0, 1])
    sns.violinplot(x="country_full", y="rank", data=paises, ax=ax)

with sns.diverging_palette(220, 20):
    ax = f.add_subplot(gs[1, 0])
    sns.violinplot(x="country_full", y="rank", data=paises, ax=ax)

with sns.color_palette("Paired"):
    ax = f.add_subplot(gs[1, 1])
    sns.violinplot(x="country_full", y="rank", data=paises, ax=ax)

f.tight_layout()
```
